```python
ba="b_1 a_2 n_3 a_4 n_5 a_6 s_7".split(" ")
```

```python
ba
```

```python
ba[0][0]
```

```python
import itertools
```

```python
baz = [x for x in itertools.permutations(ba)]
```

```python
baz
```

```python
baz[0][0][0]
```

```python
w = [q for q in baz if 
    q[0][0] == "b" and
    q[1][0] == "a" and
    q[2][0] == "n" and
    q[3][0] == "a" and 
    q[4][0] == "n" and
    q[5][0] == "a" and
    q[6][0] == "s"
]
```

```python
[" ".join(x) for x in w]
```
